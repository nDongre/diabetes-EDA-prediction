```python
# Out of all these zero values glucose, bloodpressure, BMI cannot be zero. So, let us replace them with median values
#Skin thickness being as low as zero means the person is severly underweight and that can be a contributing factor to diabetes.
#Pregnancies can always be zero.
#Insuin values being zero is an indicator of Type-1 diabetes.
```

```python
data_copy = data.copy(deep = True)
data_copy[['Glucose','BloodPressure','BMI']] = data_copy[['Glucose','BloodPressure','BMI']].replace(0,np.NaN)

## showing the count of Nans
print(data_copy.isnull().sum())
```

```python
p = data.hist(figsize = (20,20))
```

```python
#Replacig the values
data_copy['Glucose'].fillna(data_copy['Glucose'].median(), inplace = True)
data_copy['BloodPressure'].fillna(data_copy['BloodPressure'].median(), inplace = True)
data_copy['BMI'].fillna(data_copy['BMI'].median(), inplace = True)
```

```python
#Let build a model describing the relationship of our data
```

```python
#Splitting data into X and Y vesctors.
#X consists of predictor parameteres and Y consists of outcomes(predictions)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X =  pd.DataFrame(sc_X.fit_transform(data_copy.drop(["Outcome","SkinThickness"],axis = 1),),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X.head()
```

```python
Y = data_copy.drop(columns = ['Pregnancies','Glucose', 'BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
```

```python
Y.head()
```

```python
#Now let us split the dataset into training and test sets
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.10,random_state=42, stratify=Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train,Y_train.values.ravel())
```

```python
train_scores = neigh.score(X_train,Y_train)
print(train_scores)
test_scores = neigh.score(X_test,Y_test)
print(test_scores)
```

```python
#Lets draw a confusion matrix.
```

```python
from sklearn.metrics import confusion_matrix
#let us get the predictions using the classifier we had fit above
y_pred = neigh.predict(X_test)
confusion_matrix(Y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = neigh.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_proba)
```
